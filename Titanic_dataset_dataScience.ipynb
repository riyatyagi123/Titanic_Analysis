# Q10= Investigate the relationship between age and survived. Are young passengers more likely to survived

```python
plt.figure(figsize=(10,4))
sns.pointplot(data=df, x='survived',y = 'age',errorbar=('ci', False))
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=df, x='survived',y = 'age',errorbar=('ci', False))
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=df, x='survived',y = 'age')
```
